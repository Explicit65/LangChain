#### Openai Embedding Techniques

Converting Text into Vectors

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "Welcome to Appliso class on OpenAI Embeddings"
query_result = embeddings.embed_query(text) # convert the text into vectors
query_result
```

```python
len(query_result)
```

```python
embedding_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
embedding_1024
```

```python
text = "Welcome to Appliso class on OpenAI Embeddings"
query_result_1024 = embedding_1024.embed_query(text) # convert the text into vectors
query_result_1024
```

```python
len(query_result_1024)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("./data/speech.txt")
text_docs = loader.load()
text_docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(text_docs)
final_documents
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents,embedding=embedding_1024)
db
```

```python
# Retrieve the embedding and vector storeDB
query = "WHat's the speaker future ambition?"
retrived_result = db.similarity_search(query)
print(retrived_result)
```
